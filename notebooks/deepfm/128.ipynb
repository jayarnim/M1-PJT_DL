# Packages

```python
import pandas as pd
from ctr import config, featuremap, datamodule, trainer, evaluator, utils
import fm
```

```python
PATH = f"./config/{MODEL_CLS}.yaml"
cfg = config.build_config(PATH)
```

```python
utils.seed.main(cfg.seed)
```

# Load Data

```python
PATH = "./data/criteo.csv"
df = pd.read_csv(PATH)
```

```python
y_col = "label"
X_col = df.columns.difference([y_col])
```

```python
NUMERIC_COL = [
    col 
    for col in X_col 
    if "I" in col
]

CATEGORICAL_COL = [
    col 
    for col in X_col 
    if "C" in col
]
```

```python
y = df[y_col]
X = df[X_col]
```

# Data Preparation

```python
kwargs = dict(
    X=X,
    y=y,
    cfg=cfg.datamodule,
    scaling=NUMERIC_COL,
)

dataloaders = datamodule.build_datamodule(**kwargs)
```

```python
kwargs = dict(
    X=X,
    cat_cols=CATEGORICAL_COL,
    num_cols=NUMERIC_COL,
)

feature_map = featuremap.build_feature_map(**kwargs)
```

# Modeling

```python
NUM_FEATURES = 39

i = 0
units = list()

while True:
    UNIT = (2**i)
    if EMBEDDING_DIM*NUM_FEATURES < UNIT:
        break
    else:
        units.append(UNIT)
        i+=1

HIDDEN_DIM = units[:-4:-1]
```

```python
kwargs = dict(
    **cfg.model.params,
    embedding_dim=EMBEDDING_DIM,
    hidden_dim=HIDDEN_DIM,
    feature_map=feature_map,
)

model = fm.deepfm.DeepFactorizationMachine(**kwargs)
```

# Training

```python
kwargs = dict(
    model=model,
    cfg=cfg.trainer,
)

opt = trainer.build_trainer(**kwargs)
```

```python
kwargs = dict(
    trn_loader=dataloaders["trn"],
    val_loader=dataloaders["val"],
)

records = opt.fit(**kwargs)
```

```python
SUPTITLE = f"DeepFM Training History (Emb: {EMBEDDING_DIM})"
TRN_SCORE = "BCE"
VAL_SCORE = "AUROC"
FIGSIZE = (7,3)

kwargs = dict(
    records=records,
    suptitle=SUPTITLE,
    trn=TRN_SCORE,
    val=VAL_SCORE,
    figsize=FIGSIZE,
)

utils.visualizer.main(**kwargs)
```

# Evaluation

```python
kwargs = dict(
    model=model, 
    threshold=records["threshold"],
)

msr = evaluator.build_evaluator(**kwargs)
```

```python
result, metrics_sheet = msr(dataloaders["tst"])
```
