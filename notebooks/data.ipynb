# Packages

```python
from huggingface_hub import hf_hub_download
import zipfile
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

# Load Data

```python
zip_path = hf_hub_download(
    repo_id="reczoo/Criteo_x1",
    repo_type="dataset",
    filename="Criteo_x1.zip",
)
```

```python
with zipfile.ZipFile(zip_path) as z:
    with z.open("train.csv") as f:
        origin = pd.read_csv(f)
```

```python
origin.head()
```

```python
TOTAL = len(origin)
CLICK = len(origin[origin['label']==1])
NON_CLICK = len(origin[origin['label']==0])
RATIO = NON_CLICK/CLICK

print(
    f"total:\t\t{TOTAL}",
    f"click:\t\t{CLICK}",
    f"non-click:\t{NON_CLICK}",
    f"ratio:\t\t{RATIO:.4f}",
    sep="\n",
)
```

```python
y_col = "label"

df, _ = train_test_split(
    origin,
    train_size=1_000_000,
    shuffle=True,
    stratify=origin[y_col],
    random_state=42,
)
```

```python
TOTAL = len(df)
CLICK = len(df[df['label']==1])
NON_CLICK = len(df[df['label']==0])
RATIO = NON_CLICK/CLICK

print(
    f"total:\t\t{TOTAL}",
    f"click:\t\t{CLICK}",
    f"non-click:\t{NON_CLICK}",
    f"ratio:\t\t{RATIO:.4f}",
    sep="\n",
)
```

```python
X_col = df.columns.difference([y_col])

CATEGORICAL_COL = [
    col 
    for col in X_col 
    if "C" in col
]

for col in CATEGORICAL_COL:
    encoder = LabelEncoder()
    df[col] = encoder.fit_transform(df[col])
```

# Save

```python
PATH = "./data/criteo.csv"
df.to_csv(PATH, index=False)
```
